```python
import pandas as pd

# Load the dataset
df_archived = pd.read_csv("Actual Datasets/Archived data- National Charge-point Registry - Nov 2024.csv")

# Define postcode prefixes for each borough
borough_postcodes = {
    "camden": ["NW1", "WC1", "WC2"],
    "westminster": ["W1", "SW1", "WC1", "WC2"],
    "islington": ["N1", "EC1", "EC2"],
    "kensington and chelsea": ["W8", "SW7", "SW5"],
    "hammersmith and fulham": ["W6", "SW6", "W14"],
    "hackney": ["E8", "E9", "N1"],
    "tower hamlets": ["E1", "E2", "E3", "E14"],
    "lambeth": ["SE1", "SW2", "SW4", "SW9"],
    "city of london": ["EC1", "EC2", "EC3", "EC4"]
}

# Extract and clean postcode prefix
df_archived['postcode'] = df_archived['postcode'].astype(str).str.upper().str.strip()
df_archived['postcode_prefix'] = df_archived['postcode'].str.extract(r'^([A-Z]{1,2}\d{1,2})', expand=False)
df_archived['postcode_prefix'] = df_archived['postcode_prefix'].str.upper()

# Function to map postcode prefix to borough
def get_borough(prefix):
    for borough, prefixes in borough_postcodes.items():
        if prefix in prefixes:
            return borough
    return None

# Apply function and filter rows
df_archived['borough'] = df_archived['postcode_prefix'].apply(get_borough)
filtered = df_archived[df_archived['borough'].notnull()].copy()

# Drop postcode_prefix if you no longer need it
filtered = filtered.drop(columns=['postcode_prefix'])

# Save the result
filtered.to_csv("Actual Datasets/chargingStations.csv", index=False)
```

```python
# Get value counts of device networks from the filtered dataset
device_network_counts = (
    filtered['deviceNetworks']
    .fillna("Unknown")
    .value_counts()
    .rename_axis("device_network")
    .reset_index(name="num_chargers")
)
device_network_counts
```
